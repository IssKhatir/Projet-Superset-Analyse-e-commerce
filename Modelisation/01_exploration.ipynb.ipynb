# Analysis of global sales

## 1. Sales volume & CA

```python
# Total sales volume
print("Total des produits vendus :", df['Quantity'].sum())

# Chiffre d'affaires total
df['TotalPrice'] = df['Quantity'] * df['Price']
print("Chiffre d'affaires total :", df['TotalPrice'].sum(), "£")
```

## 2. Monthly distribution

```python
df['Month'] = df['InvoiceDate'].dt.to_period('M').dt.to_timestamp()

monthly_sales = df.groupby('Month').agg({
    'Quantity': 'sum',
    'TotalPrice': 'sum'
}).reset_index()
monthly_sales
```

```python
plt.figure(figsize=(12,5))
sns.lineplot(data=monthly_sales, x='Month', y='TotalPrice', marker='o')
plt.title("CA mensuel")
plt.xticks(rotation=45)
plt.ylabel("TotalPrice (£)")
plt.show()
```

## 3. Distribution by day of the week

```python
df['Weekday'] = df['InvoiceDate'].dt.day_name()

weekday_sales = df.groupby('Weekday')['TotalPrice'].sum().reindex([
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
])
weekday_sales
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x=weekday_sales.index, y=weekday_sales.values)
plt.title("CA per day of the week")
plt.ylabel("CA (£)")
plt.xticks(rotation=45)
plt.show()
```

## 4. Distribution by hour

```python
df['Hour'] = df['InvoiceDate'].dt.hour

hourly_sales = df.groupby('Hour')['TotalPrice'].sum()
hourly_sales
```

```python
plt.figure(figsize=(10,4))
sns.lineplot(x=hourly_sales.index, y=hourly_sales.values, marker='o')
plt.title("CA per hour of the day")
plt.xlabel("Heure")
plt.ylabel("CA (£)")
plt.grid()
plt.show()
```

# Visualisations

## 1. Total sales by country

```python
country_sales = df.groupby("Country")["TotalPrice"].sum().sort_values(ascending=False).reset_index()
country_sales
```

```python
# Barplot
plt.figure(figsize=(18,6))
sns.barplot(data=country_sales, x="Country", y="TotalPrice", palette="viridis")
plt.xticks(rotation=45)
plt.title("Total sales by country")
plt.tight_layout()
plt.show()
```

## 2. Top products by CA

```python
top_products = df.groupby("Description")["TotalPrice"].sum().sort_values(ascending=False).head(10).reset_index()
top_products
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=top_products, y="Description", x="TotalPrice", palette="rocket")
plt.title("Top 10 products by revenue")
plt.xlabel("Total sales")
plt.ylabel("Products")
plt.tight_layout()
plt.show()
```

## 3. Heat map (heatmap) product country (aggregated)

```python
# a country x product matrix with the sum of sales
top_countries = df.groupby("Country")["TotalPrice"].sum().nlargest(5).index
top_products = df.groupby("Description")["TotalPrice"].sum().nlargest(10).index

filtered = df[df["Country"].isin(top_countries) & df["Description"].isin(top_products)]
heatmap_data = filtered.pivot_table(index="Country", columns="Description", values="TotalPrice", aggfunc="sum", fill_value=0)
heatmap_data.head(5)
```

```python
plt.figure(figsize=(18,10))
sns.heatmap(heatmap_data, cmap="YlGnBu", cbar_kws={"label": "Total Sales"})
plt.title("Heatmap: Sales by country and product")
plt.xlabel("Product")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```
